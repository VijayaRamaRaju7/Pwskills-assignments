```python
# 3. Given a recurrence relation, solve it using the recursive tree approach:
# a. T(n) = 2T(n-1) +1
# b. T(n) = 2T(n/2) + n


a.T(n) = 2T(n-1) +1
recursive tree for T(n) = 2T(n-1) +1


                        T(n)
                       /    \
                 T(n-1)    T(n-1)
                  / \       / \
            T(n-2) T(n-2) T(n-2) T(n-2)
             / \    / \    / \    / \
         T(n-3) T(n-3) ...  ...  T(n-3)  (continues for a total of n levels)
                
Level 1: 1 node (T(n))
Level 2: 2 nodes (T(n-1))
Level 3: 4 nodes (T(n-2))
...
Level k: 2^k nodes (T(n-k))
    
    
The total number of nodes in the tree is 1+2+3+4...+2^n-1  which is a geometric series. The sum of a geometric series with a=1,r=2 and n terms is given by

S=a((r^n)/(r-1))

In this case S = 2^n-1

Therefore, the time complexity is O(2^n)



b. T(n) = 2T(n/2) + n

T(n) = 2T(n/2) + n

recursive tree for T(n) = 2T(n/2) + n

                        T(n)
                       /    \
                 T(n/2)    T(n/2)
                  / \       / \
              T(n/4) T(n/4) T(n/4) T(n/4)
               / \    / \    / \    / \
         T(n/8) T(n/8) ...  ...  T(n/8)  (continues for a total of log(n) levels)
                

                
Level 1: 1 node (T(n))
Level 2: 2 nodes (T(n/2))
Level 3: 4 nodes (T(n/4))
...
Level k: 2^k nodes (T(n/k))
    
    
The total number of nodes in the tree is 1+2+3+4...+2 log2n which is a geometric series. The sum of a geometric series with a=1,r=2 and n terms is given by

S=a((r^n)/(r-1))

In this case S = n-1+n=2n-1

Therefore, the time complexity is O(n)
```
